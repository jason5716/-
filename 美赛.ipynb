#合作伙伴可以用jointplot算相关关系来进一步验证合作伙伴
#数据要用每日收盘价的比率

```python
closing_df=pd.DataFrame({'byd':byd['close'],'zsy':zsy['close']})
closing_df.head()
```

```python
auto_rets=closing_df.pct_change()
auto_rets.head()
```

```python
with sns.axes_style('whitegrid'):
    h=sns.jointplot('byd','zsy',auto_rets,kind='scatter',color='seagreen',stat_func=sci.pearsonr, size=6, ratio=5, space=0.2, dropna=True,marginal_kws = dict(bins=15, rug =True))
    h.set_axis_labels('PetroChina', 'byd', fontsize=13)
#sns.jointplot('byd','zsy',auto_rets,kind='scatter')
```

```python
closing_df2=pd.DataFrame({'gudong4':gudong4['close']})
tech_rets2=closing_df2.pct_change()
tech_rets2.head()
```

```python
rets = tech_rets2.dropna()

# Scatter plot of expected return of the stocks vs. their standard deviations of daily retunrs
area = np.pi*20

plt.scatter(rets.mean(), rets.std(),alpha = 0.5,s =area)

plt.ylim([-0.01,0.025])
plt.xlim([-0.005,0.001])

plt.xlabel('Expected returns')
plt.ylabel('Risk')
plt.grid(ls=':',c='gray')

for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(label, 
        xy = (x, y), xytext = (50, 30),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        arrowprops = dict(arrowstyle = '-', connectionstyle = 'arc3,rad=0.5',color='red'))
```
